```python
# Definindo os caminhos dos arquivos CSV de exportação e importação
file_path = '/content/drive/MyDrive/df_agro_final_2.xlsx'
```

```python
df = pd.read_excel(file_path)
```

```python
df_exp_2024 = df.copy()
```

```python
# Agrupando os dados de exportação por NCM somando os valores FOB
df_exp_2024 = df_exp_2024.groupby(['ncm', 'origem', 'mes', 'pais_importador', 'estado_produtor', 'tipo'], as_index=False)['valor_total'].sum()
```

```python
df_exp_2024.head()
```

```python
# Calculando o valor total geral
total_exp_2024 = df_exp_2024['valor_total'].sum()

# Criando uma nova coluna com a participação percentual de cada produto
df_exp_2024['PERC_PARTICIPACAO_EXP'] = ((df_exp_2024['valor_total'] / total_exp_2024) * 100).round(2)
```

```python
# Salvando o DataFrame em um arquivo CSV
df_exp_2024.to_csv('dados_preparados.csv', index=False)
```
